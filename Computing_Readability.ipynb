```python
def concatenate_columns(row, columns_to_concatenate):
    concatenated_text = ''
    for column in columns_to_concatenate:
        if pd.notnull(row[column]):
            concatenated_text += str(row[column]) + ' '
    return concatenated_text.strip()
```

```python
columns_to_concatenate = ['prompt', 'lrn_option_0', 'lrn_option_1', 'lrn_option_2', 'lrn_option_3', 'lrn_option_4', 'lrn_option_5', 'lrn_option_6', 'lrn_option_7', 'lrn_option_8', 'lrn_option_9', 'lrn_option_10', 'lrn_option_11']
new_column_name = 'text'

q_bank[new_column_name] = q_bank.apply(lambda row: concatenate_columns(row, columns_to_concatenate), axis=1)
```

```python
import re

def clean_text(text):
    words_only = re.findall(r'\b\w+\b', text)
    sentence = ' '.join(words_only)
    return sentence

def clean_dataframe(df, columns):
    # Iterate over each column and apply text cleaning
    df[columns] = df[columns].apply(clean_text)
    return df
```

```python
cleaned_df = clean_dataframe(q_bank, 'text')
cleaned_df.to_csv('cleaned_df.csv')
```

```python
import nltk
import pandas as pd

# Download the word list from NLTK
nltk.download('words')

# Function to load the dictionary
def load_dictionary():
    from nltk.corpus import words
    english_words = set(words.words())
    return english_words

# Function to count actual words in the string based on the dictionary
def count_actual_words_in_dictionary(string, dictionary):
    words = string.split()
    count = 0
    valid_words = []
    for word in words:
        if word.lower() in dictionary:
            count += 1
            valid_words.append(word)
    return count, ' '.join(valid_words)

# Load the dictionary
dictionary = load_dictionary()


# Apply the function to count actual words in each string
cleaned_df['actual_word_count'], cleaned_df['valid_words'] = zip(*cleaned_df['text'].apply(lambda x: count_actual_words_in_dictionary(x, dictionary)))

print(cleaned_df)
```

```python
cleaned_df['valid_words']
```

```python
def calculate_readability_scores(text):
    flesch_reading_ease = textstat.flesch_reading_ease(text)
    flesch_kincaid_grade = textstat.flesch_kincaid_grade(text)
    dale_chall_readability_score = textstat.dale_chall_readability_score(text)
    gunning_fog = textstat.gunning_fog(text)
    coleman_liau_index = textstat.coleman_liau_index(text)
    smog_index = textstat.smog_index(text)
    automated_readability_index = textstat.automated_readability_index(text)

    return {
        "Flesch Reading Ease": flesch_reading_ease,
        "Flesch-Kincaid Grade": flesch_kincaid_grade,
        "Dale-Chall Readability Score": dale_chall_readability_score,
        "Gunning Fog": gunning_fog,
        "Coleman-Liau Index": coleman_liau_index,
        "SMOG Index": smog_index,
        "Automated Readability Index": automated_readability_index
    }

cleaned_df['Readability Scores'] = cleaned_df['valid_words'].apply(calculate_readability_scores)
readability_scores = cleaned_df['valid_words'].apply(calculate_readability_scores).apply(pd.Series)
q_new=pd.concat([cleaned_df, readability_scores], axis=1)
q_new.to_csv('text_scores.csv')
```
